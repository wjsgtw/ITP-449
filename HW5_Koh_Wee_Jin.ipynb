```python
# import all necessary libraries here

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb
from sklearn import metrics
from sklearn.metrics import plot_confusion_matrix
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
```

```python
# Type solution here
#1
df = pd.read_csv('Titanic.csv')
df.head()
```

```python
#2
print(df.head())
print(df.shape)
print(df.columns.values)

#The column/variable that is most useful as a target variable is Survived because survival is the end outcome of the each passenger's ordeal.
```

```python
#3
#Passenger column is not important for logistic regression and is the same as index, so drop it.
df.drop('Passenger', axis=1, inplace=True)

#Class is important since the lower the class of travel, the farther one is likely to be from lifeboats and can affect survivability.
#Within Class, Crew members also have a professional and ethical obligation to remain with the ship as long as possible, and can affect survivability.

#Sex and Age are also important, since women and children are given priority on lifeboats and can affect survivability.

#Therefore, Class, Sex, and Age need to remain in addition to target variable Survived
```

```python
#4
print(df.isnull().sum())

#Since there are no missing values, no need to handle missing values
```

```python
#5
#Plot count plots of each of the remaining factors (including the target variable)
plt.figure(1)
sb.countplot(x='Class', data=df)

plt.figure(2)
sb.countplot(y='Sex', data=df)

plt.figure(3)
sb.countplot(y='Age', data=df)

plt.figure(4)
sb.countplot(x='Survived', data=df)

plt.show()
```

```python
#6
#create dummy variables for the remaining attributes other than the target variable
df2 = pd.get_dummies(df, columns=['Class', 'Sex', 'Age'])
df2.head()
```

```python
#7
#Split the data into a training and test set.

X = df2.iloc[:,1:]
y = df2.iloc[:,0]

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=2020)
print(X_train.shape)
print(y_train.shape)
print(X_test.shape)
print(y_test.shape)
```

```python
#8
#Logistic Regression model
model = LogisticRegression()

model.fit(X_train, y_train)
```

```python
#9
#predict outcomes for test set
y_pred = model.predict(X_test)
print(y_pred)
```

```python
#10
#accuracy of prediction for survivability
print(metrics.accuracy_score(y_test, y_pred))
```

```python
#11
#Plot and print confusion matrix
labels = ['No', 'Yes']
cnf_matrix = metrics.confusion_matrix(y_test, y_pred)
print(cnf_matrix)

metrics.plot_confusion_matrix(model, X_test, y_test)

#display with Yes, No labels
plt.xticks(np.arange(len(labels)),labels=labels)
plt.yticks(np.arange(len(labels)),labels=labels)
plt.xlabel("Predicted Survival")
plt.ylabel("True Survival")
plt.show()
```

```python
#12
#display the predicted value of the survivability of a male adult passenger traveling in 3rd class

#create sample dataframe row for adult male in 3rd class, and predict value through model
thirdClassMaleAdult = [[0,0,1,0,0,1,1,0]]
pred = model.predict(thirdClassMaleAdult)
print(pred)
```

```python
#13
#display the predicted probability of the survivability of a male adult passenger traveling in 3rd class.

#predict probability through model, and taking the second column only since that is the probability of survival.
predProb = model.predict_proba(thirdClassMaleAdult)[:,1]
print(predProb)
```
